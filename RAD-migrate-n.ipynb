## The model  
I simulate data under the following equilibrium scenario. Five populations diverged as a polytomy. Since their divergence 1.0 coalescent time (2N) units in the past, popA and popB have exchanged 50 migrants per generation, while popC and popD have exchanged 10 migrants per generation. The other populations, popE, has remained isolated from gene flow. 

![](https://dl.dropboxusercontent.com/u/2538935/PyRAD_TUTORIALS/tutorial_migrate/drawing.png)

### Parameters used by `migrate-n` and `simLoci` 

From Peter Beerli's website:  

    "migrate expresses its parameters scaled by the mutation rate because when using genetic data from a single time point we are not able to separate the influence of mutation µ (mu) and population size, thus we use the product of the two and have the mutation-scaled effective population size [theta] which is 4Neµ and the mutation-scaled immigration rate M which is m/µ."

```python
m1 = 1e-7     ## migration rate
m2 = 2e-8     ## migration rate
Ne = 1e6      ## population size
x = 4         ## diploids inheritance
u = 2e-9      ## per site mutation rate


print "Mutation-scaled parameters to estimate:"
## per site theta (the mutation-scaled population size)
theta = x*Ne*u
print theta, "\t theta"

## migration (the mutation-scaled immigration rate)
M1 = m1/u
print M1, "\t M1"
M2 = m2/u
print M2, "\t M2\n"

## check that all is well
print "x*Ne*m1 == theta*M1:", x*Ne*m1 == theta*M1
print "x*Ne*m2 == theta*M2:", x*Ne*m2 == theta*M2

## print 
print "\nNumber of diploid migrants per gen (theta*M):"
print theta*M1  
print theta*M2
```

### Simulate data in `simLoci`. 
The arguments passed are shown below. 

```python
%%bash
python simLoci.py
```

```python
%%bash
## save the example input tree as a newick string
## resolve polytomies with nearly zero length branch lengths
## set branch lengths of each lineage at 1.0 coalescent units (2Ne)
tree="(((A:1.0,B:1.0):0.0001,(C:1.0,D:1.0):0.0001):0.0001,E:1.0002);"

## save each migration scenarios as a string, separated by "/"
## 50 migrants per gen between A-B
## 10 migrant per gen between C-D
## no migration w/ E
## [recipient,donor,time-end,time-start,migration-rate]
migs="\
[A,B,0.0,1.0,1.0e-7]/\
[B,A,0.0,1.0,1.0e-7]/\
[C,D,0.0,1.0,2.0e-8]/\
[D,C,0.0,1.0,2.0e-8]"

## simulate data with the following parameters
python simLoci.py  -L 50 \
                   -l 100 \
                   -u 2e-9 \
                   -N 1e6 \
                   -i 10 \
                   -t $tree \
                   -m $migs \
                   -s 54321 \
                   -f m \
                   -o "sim50"
```
