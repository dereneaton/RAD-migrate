## The model  
I simulate data under the following equilibrium scenario. Five populations diverged as a polytomy. Since their divergence many (20) coalescent time units (2N) in the past, A and B have exchanged migrants unidirectionaly, while C and D have exchanged migrants bidirectionaly, but at different rates. Population E has remained isolated from gene flow. The arrows on the migration matrix are showing migration forward in time, as opposed to how they are inferred by migrate as backwards in time (e.g., m$_{C->D}$ in migrate means migration from D into C forward in time, or C into D thinking backwards as in the coalescent).

![](https://dl.dropboxusercontent.com/u/2538935/PyRAD_TUTORIALS/tutorial_migrate/drawing.png)

### Parameters used by `migrate-n` and `simLoci` 

From Peter Beerli's website:  

    "migrate expresses its parameters scaled by the mutation rate because when using genetic data from a single time point we are not able to separate the influence of mutation µ (mu) and population size, thus we use the product of the two and have the mutation-scaled effective population size [theta] which is 4Neµ and the mutation-scaled immigration rate M which is m/µ."

```python
## My simulation parameters
mAB = 2e-6     ## migration rate
mCD = 2e-7     ## migration rate
mDC = 1e-7     ## migration rate
Ne = 1e5        ## Eff. population size
x = 4           ## diploids inheritance
u = 1e-8        ## per site mutation rate


## per site theta (the mutation-scaled population size)
print "Mutation-scaled parameters to estimate:"
theta = x*Ne*u
print 'theta\t', theta/4, "("+`theta`+")"


## migration (the mutation-scaled immigration rate)
MAB = mAB/u
MCD = mCD/u
MDC = mDC/u
print "MAB\t",MAB
print "MCD\t",MCD
print "MDC\t",MDC


## thinking as individuals movement 
print "\nNumber of diploid migrants per gen (theta*M):"
print theta*MAB
print theta*MCD
print theta*MDC
```

### Simulate data in `simLoci`. 
The arguments passed are shown below. 

```python
%%bash
python simLoci.py
```

```python
%%bash

## save the example input tree as a newick string
## resolve polytomies with nearly zero length branch lengths
## set branch lengths of each lineage at xx coalescent units (2Ne)
## deep divergence ensures stationarity (assumed equilibrium)
## and less ILS to bugger up estimates of migration
tree="(((A:20.0,B:20.0):0.0001,(C:20.0,D:20.0):0.0001):0.0001,E:20.0002);"

## save each migration scenarios as a string, each separated by "/"
## migration rates are per generation.
## input order backwards time thinking: [donor,recipient,time-end,time-start,migration-rate]
## input order forward time thinking: [recipient,donor,time-start,time-end,migration-rate]
## Time 0 is the present, so the backward time thinking may be more clear
## The scenario described in the figure above is designated below
migs="\
[A,B,0,20,1e-6]/\
[C,D,0,20,1e-7]/\
[D,C,0,20,2e-7]"

## simulate data with the following parameters
python simLoci.py  -L 200 \
                   -l 100 \
                   -u 1e-8 \
                   -N 1e5 \
                   -i 10 \
                   -t $tree \
                   -m $migs \
                   -s 54321 \
                   -f m \
                   -o "sim200"
```
